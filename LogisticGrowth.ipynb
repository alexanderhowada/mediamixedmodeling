```python
import numpy as np

import plotly.graph_objects as go
```

```python
class LogisticGrowth:
    def __init__(self, scale, offset):
        self.scale = scale
        self.offset = offset
        
    def __call__(self, x):
        return 1/(1+np.exp(-(x-self.offset)/self.scale))
```

```python
if __name__ == '__main__':
    x = np.arange(100)
    
    fig = go.Figure()
    offset = 40
    for scale in [1, 2, 5, 10, 20]:
        LG = LogisticGrowth(scale, offset)
        y = LG(x)
        
        fig.add_trace(
            go.Scatter(x = x, y = y, name = scale)
        )
    fig.update_layout(title=dict(text=f'offset={offset}', x=0.5), legend=dict(title='scale'))
    fig.show()
    
    fig = go.Figure()
    scale = 5
    for offset in np.arange(0, 70, 15):
        LG = LogisticGrowth(scale, offset)
        y = LG(x)
        
        fig.add_trace(
            go.Scatter(x = x, y = y, name = scale)
        )
    fig.update_layout(title=dict(text=f'scale={scale}', x=0.5), legend=dict(title='offset'))
    fig.show()
```
